# 1.1 More indexing (MASKING)
*What if we only want to look a rows which satisfy a condition?*

For example only the large flowers. Or only the virginica flowers. Could we ask for a few particular rows of the `dataframe df`?

Just like `df['petal_length']` is a *Series* (single column of dataframe) `df['petal_length'] > 5` is also a series of `True` and `False` - True and False as a data type are called *booleans* 

````
df['petal_length'] > 5
````

Having a *Series* (single column of dataframe) of just `True` and `False` might not seem very useful... but lets assign it to **a new variable name** called `mask`

````
mask = df['petal_length'] > 5
````

````
mask 
````

The key is being able to **INDEX** with this mask (a series of booleans) on top of the original dataframe `df`. Instead of getting all 150 rows **we will only get the rows of the dataframe where** `mask` **has a value of** `True` 





````
df[mask]
````

>Essentially, we are only looking at the flowers in the dataset where the petal length is greater than 5. 
We have another dataframe here, lets assign it to the variable `df_large`. 
Note that 


```
mask = df['petal_length'] > 5
df[mask]
```
***is the same as***
```
df[df['petal_length'] > 5]
```
*We introduced it the first way to make things easier to understand, avoiding the double* `[[ ]]`


````
df_large = df[df['petal_length'] > 5]
````

And we inspect: 

````
df_large
````

Looking at `df_large` again, we can see that all but one of the flowers within are of the `species` `virginica`. This is our first insight in the data, larger flowers (based on `petal_length`) are more likely to be `virginica`

# 1.2 Easy Plots (SEABORN) 
*Is there an easier way to do all this?*

The easiest way to plot our data is to use a library called seaborn (imported earlier: `import seaborn as sns`) and its plotting functions. These functions ONLY WORK WITH data in a `dataframe`, so if you wouldn't be able to plot a list of numbers with it. 




Seaborn has many functions for many plots, in each you pass in the dataframe you want to plot with `data = df`. 

````
sns.pairplot(data = df)
````

`sns.pairplot()` is a very powerful function which plots every number column against each other. There are 14 subplots, eg the bottom left plots the `'petal_width' vs 'sepal_length'` of all 150 flowers, while the diagonals show a histogram (more on that later!) 

As a function, you can ***HOVER OVER WITH YOUR MOUSE TO LEARN MORE*** The `data=` input is required, we also see the  `hue = ` input, you can set it to **any of the columns** in the data and the plot will automatically adjust colours to represent more information.

Here we use `hue = 'species'`:

````
sns.pairplot(data = df, hue = 'species')
````

Use `sns.pairplot()` with caution, as it can take too long for large `df`s!

In our dataframe: `species` is a **CATEGORY** (it always takes one of three distinct values) while ***ALL THE OTHER COLUMNS*** are ***CONTINOUS*** (it can take any value on a number line)

````
df
````

Generally, we want to put a **CATEGORY** for color (`hue = `) and plot numbers/continuous values on the `x=` and `y=` (as you are probabily familiar with in Math class) Here is another plot to demonstrate:

````
sns.scatterplot(data = df, x = 'petal_length', y = 'petal_width', hue = 'species')
````

# 1.3 Basic Plots (MATPLOTLIB)
*making a basic plot*

Seaborn is built on top of matplotlib (which we imported as `plt`), and is limited to `dataframes` ONLY. So what if we wanted to plot a list of numbers? Then we shall use matplotlib!!

Matplotlib takes a few more lines of code to set up: every plot in matplotlib starts with the same ***TEMPLATE*** like so:





````
fig = plt.figure()
ax = plt.axes()


fig.show()
````

`fig` and `ax` are called actually classes, which will be explained by the end of today. For now think of `fig` as an empty canvas which you have to *show* at the end, and `ax` as the grid which you will plot on. We will *copy* this everytime we make a plot. 


We'll start with the `ax.scatter()` function. It comes from the plt (matplotlib) library and makes scatter plots. **REMEMBER THE HOVER**, as we hover over the text `ax.scatter()`, we should see 

* "A scatter plot of *y* vs. *x* with varying marker size and/or color." 


**Keep hovering** On the top are the inputs the function needs, anything that has an equal sign `=` has a default value which we don't care about. The first two inputs `x` and `y` must be supplied by us. It says `x` and `y` must be array like, a `list` is an array so lets put that in. 

````
fig = plt.figure()
ax = plt.axes()

ax.scatter(x = [1,2,3,4] , y = [1,4,9,16]) 

fig.show()
````

We used a single equal sign but we did not actually assign `x= [1,2,3,4]`. When the equal sign is in function, it is simply telling the function to use `[1,2,3,4]` internally to plot (it's a plotting function). `x` is not a variable we can use later.  

This is not a good plot, it will score 3/10 in any Engineering report. lets make some adjustments. 




````
fig = plt.figure()
ax = plt.axes()

ax.scatter(x = [1,2,3,4], y = [1,4,9,16])
ax.set(xlabel = 'x values', ylabel = 'y values', title = 'Simple Plot') 

fig.show()
````

So far we have
1.  We make a figure called `fig`, acts like an empty canvas
2.  We make an axes called `ax`, acts like a grid on the figure
3.  We plot on this axes using `ax.scatter()` same thing as `plt.scatter()`
5. We set some parameters on the axes using `ax.set()`
4. We show the figure

If you hover over `ax.set()` you might notice that it just asks for **kwargs, think of it as there being so many inputs we can't list all of them here. You just have to know which inputs you need, or refer to the list [here](https://matplotlib.org/3.1.1/api/axes_api.html)

Another useful type of plot is a histogram, learn more about histograms [here](https://www.mathsisfun.com/data/histograms.html). This is done using the `ax.hist()` function, for this function all you need are the x values. 

Also, note that:  ` fig, ax = plt.subplots() ` serves the exact same purpose as: 
```
fig = plt.figure()
ax = plt.axes()
```
so we will use this from now on to save time. (Essentially shortening the plotting template.)



````
fig, ax = plt.subplots()

ax.hist(x = df['petal_length'])
ax.set(xlabel = 'length [cm]', title = 'Histogram of Petal Length')

fig.show()
````

The above plot shows the number of flowers (or the *count*) which have a `petal_length` between (0,1), (1,1.5), (1.5, 2) etc. Thus showing a *distribution* of the column of numbers, which are ***CONTINUOUS** You can also make a histogram of a **CATEGORY**, such as `'species'`

````
fig, ax = plt.subplots()
ax.hist(x = df['species'])
ax.set(xlabel = 'Type of Flowers', title = 'Number of flowers in eacy category')
fig.show()
````

#### OBJECTS in PYTHON (to be discussed in detail next class!)

First lets talk about all the variables we have created so far. We have a few different `df` dataframes which are all storing different information. We have `fig`, and `ax` we have been overwriting each time. All of these are considered `objects`. In fact, everything in Python is an object.  

These objects all come from the libraries we imported at the very top. These objects all have **many functions** which you can use, sometimes they can be **indexed** and of course they can **store information** for you. 

I like to think of these objects as Robots. Each robot having its own *functions*. The first robot we made make is from `plt.figure()`, and we'll call this robot `fig`. We could have called it anything we wanted to: 



````
another_figure = plt.figure()
this_axes = plt.axes()

another_figure.show()
````

Here, `another_figure` is another *instance* of the robot `plt.figure()` and is made from the "same "same manufacturer" so it has the same functions as `fig`, yet it can and will store completely different data, in this case plots/images. 

It also applies to `df` and `df_virginica` both are dataframes from the library `pandas`. And we can index them the same way, use the same functions to manipulate them (tomorrow we learn there are hundreds of functions which can be applied to a `dataframe`.) However they are still different in the information (data) they store, when we apply the same function on both, we will get different results. 

````
df.head()
````

````
df_virginica.head()
````

Okay back to plotting: we already know figure is like a blank canvas, it doesn't have many functions except `.show()` to show the figure and `.savefig()` to save it as a jpeg or png file. 

The plotting happens on another, completely different robot called `ax`
`ax` is very powerful as it comes from another class `plt.axes()` , unlike the dumb robot `fig` which only has 2 functions, `ax` has tons of functions. Our data will be plotted on the `ax` which is housed and displayed in the `fig`.  Why do this? because we can have more than on `axes` on the same `figure`. 

Now we will look at the most confusing part of DAY 1. Rather than making a `fig` and `ax` separately, the function plt.subplots(nrows = 3) will create a `figure` (singular) and a set of `axes` (plural) for us to plot on. We specify how many rows using nrows. Since we have mutiplt `axes` now, we must index `axes` using `[]` to make individual plots. 

````
fig, axes = plt.subplots(nrows = 3)

ax0 = axes[0]
ax1 = axes[1]
ax2 = axes[2]

ax0.scatter(x = df_setosa['petal_length'], y = df_setosa['petal_width'], c = 'red', label = 'setosa')
ax1.scatter(x = df_virginica['petal_length'], y = df_virginica['petal_width'], c = 'blue', label = 'virginica')
ax2.scatter(x = df_versicolor['petal_length'], y = df_versicolor['petal_width'], c = 'green', label = 'versicolor')
fig.show()
````

We can save some varible names by directly using `axes[0].scatter` and also make the figure bigger by giving `plt.subplots(nrows = 3)` a `figsize = [10,10]` value, which is the x and y dimension of the figure. 


````
fig, axes = plt.subplots(nrows = 3, figsize = [10,10])

axes[0].scatter(x = df_setosa['petal_length'], y = df_setosa['petal_width'], c = 'red', label = 'setosa')
axes[1].scatter(x = df_virginica['petal_length'], y = df_virginica['petal_width'], c = 'blue', label = 'virginica')
axes[2].scatter(x = df_versicolor['petal_length'], y = df_versicolor['petal_width'], c = 'green', label = 'versicolor')
fig.show()
````

And to customize each of the subplots, we would have to use `axes[0].set()` for each axes. And now, we want to have the legend for the entire figure rather than each axes, so we use `fig.legend()`

````
fig, axes = plt.subplots(nrows = 3, figsize = [10,10])

axes[0].scatter(x = df_setosa['petal_length'], y = df_setosa['petal_width'], c = 'red', label = 'setosa')
axes[1].scatter(x = df_virginica['petal_length'], y = df_virginica['petal_width'], c = 'blue', label = 'virginica')
axes[2].scatter(x = df_versicolor['petal_length'], y = df_versicolor['petal_width'], c = 'green', label = 'versicolor')

axes[0].set(ylabel = 'width [cm]', title = 'Plot of Petal Width vs Length of Setosa flowers only')
axes[1].set(ylabel = 'width [cm]', title = 'Plot of Petal Width vs Length of Virginica flowers only')
axes[2].set(xlabel = 'length [cm]', ylabel = 'width [cm]', title = 'Plot of Petal Width vs Length of Versicolor flowers only')
fig.legend()
fig.suptitle('Data from Iris Dataset')
fig.show()
````
