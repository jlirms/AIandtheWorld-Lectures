### Null values - such a pain...

Often the first things we do to a new dataset involves **finding and removing the null values**. Those are empty values in the data, unknowns that should be avoided when we go to modelling - as the model expects inputs and numbers, a blank(missing/null) value would just ruin things...

````
df_in.isnull()
````

`df.isnull()` function checks every single value in the dataframe and returns `True` if it is a null, as you can see most of the values are `False` - not nulls

````
nulls = df_in.isnull()
````

After we save this (entire dataframe of `True` and `False`) we can take the **SUM** - in computer speak `True = 1` and `False = 0`

````
nulls.sum()
````

This takes the sum across all columns, and we see that all our null values are in the `horsepower` column, interesting. 

We can also use the `.any()` function on `nulls` which tells us if **any** of **each row** has a True (True = null value exists!)

````
nulls.any(axis = 1)
````

This gives us a column of True and Falses, **A COLUMN of TRUE and FALSES?!?!** lets index. 

````
df_in[nulls.any(axis = 1)]
````

This is how we find the rows in our data which have null values in a dataframe. as we expected they are all in the horsepower column. 

Note that there used to be more rows in `df_in` with null values, but we manually added in some values with `df_in.loc[32, 'horsepower'] = 75`. One option to deal with nulls would be to find the missing information and fill it in one by one. 

With this newfound info, and wanting to *get rid of the null values* one option would be to completely drop '`horsepower'` or drop the rows with empty values in `horsepower`

But that seems like overkill for just 6 null values (the rest of the data in the `horsepower` column is still okay!) The other option would be to drop those specific rows or columns. We use the `df.dropna()` function. 

````
df_in.dropna(axis = 0, how = 'any')
````

The function above returns a new dataframe with the 5 rows we looked at earlier removed. **Notice how it only has 393 rows rather than 398**  Meanwhile the function use below doesn't do anything, because no row is has `all` null values. 

````
df_in.dropna(axis = 0, how = 'all')
````

The below drops columns (1 -> co1umns) with any null values, therefore `horsepower` is gone. 

````
df_in.dropna(axis = 1, how = 'any')
````

In the end `.dropna(axis = 0, how = 'any')` makes the most sense. Since it keeps all the column/features and removes all the nulls. Rightfully, this is also the default (check by hovering)

````
df_nonull = df_in.dropna(axis = 0, how = 'any')
````

# 2.4 Creating new FEATURES

As we get closer towards machine learning, we need to start treating our data. A simple example would be to multiply, divide different columns together. Or to scale certain columns, take the logarithm of others.  

We want to start with `df_nonull` since having null values will affect any mathematical operations. `df_filled` could work since it has 0s for all the null values, but we may also have division by zero problems

Lets make a copy of this dataframe which we will manipulate. This way, `df_nonull` does not change. We call it `df_input` since it will eventually be the input to our M.L. model. Note that if you had just


```
df_input = df_nonull ## this is WRONG
```
then any changes you make to `df_input` would affect `df_nonull`. The right way to make a separate copy is to use the `.copy()` function. 



> **In this section df_input will be ALTERED itself, we are no longer using functions that created modified copies**

````
df_input = df_nonull.copy()
````

For the cars dataset, what if we wanted to find the **weight to horsepower ratio**? The ratio is defined as `horsepower / weight` The symbol for division is: `/` and multiplication is: `*`

````
df_input['horsepower'] / df_input['weight']
````

This works! Now all we need to do is to declare a new column, and set its values equal to what we found. 

````
df_input['weight_to_hp_ratio'] = df_input['horsepower'] / df_input['weight']
````

Here we have made a new column and actively CHANGED `df_input`. It is now different from `df_nonull`

````
## Check for yourself, df_input and df_nonull
````

See how the new column has row numbers as well, this is what matches to the row numbers (aka index, in **bold**) of `df_input`. The index does not have all the numbers from 0 to 397 since we dropped a few rows from the `dropna()` function earlier. 

Its CRUCIAL that the index of the new column matches the index of the dataframe. 

````
df_input.head()
````

One column that we must deal with is the `origin` column. Almost every,  (basically every) ML algorithm are **mathematical** and requires **numbers as inputs.** This column only has strings: 

````
df_input['origin']
````

One option would be to make `usa = 1 europe = 2 and japan = 3` but remember that algorithms use math so doing so would imply `usa + europe = japan` since `1 + 2 = 3`. A better option would be to make three new columns, setting it to `is_usa, is_europe, is_japan`

````
df_input['origin'] == 'usa'
````

This is a **double equals sign** which is not an assignment of new values, nothing has changed we are *inspecting*, we are just comparing the columns `origins` to see if it is `usa` and we get a column (series) of True and Falses.

Now we will assign these series to new columns in the data. The parentheses here are not required, just used to make it more readable. 

````
df_input['is_usa'] = (df_input['origin'] == 'usa')
````

````
df_input['is_europe'] = (df_input['origin'] == 'europe')
df_input['is_japan'] = (df_input['origin'] == 'japan')
````

````
df_input
````

Technically speaking, `True` and `False` are not numbers, but remember inthe computer world `True = 1` and `False = 0` so the algorithm will accept this. However there is a better way of making the new columns, if the `origin` had 10 countries, this method would be tedious, we will learn this in the next chapter. But first

Fun fact, there are usually easier ways to do things in Python, we can accomplish the same task as as above by using a `pd.get_dummies` function

````
pd.get_dummies(df_nonull, columns = ['origin'])
````
